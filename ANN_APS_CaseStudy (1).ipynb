```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
import tensorflow.keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout
```

```python
classifier=Sequential()

classifier.add(Dense(units=32,kernel_initializer='he_uniform',activation='relu',input_dim =170))

classifier.add(Dense(units=16,kernel_initializer='he_uniform',activation='relu',input_dim =170))

classifier.add(Dense(units = 8, kernel_initializer = 'he_uniform',activation='relu'))

classifier.add(Dense(units = 1, kernel_initializer = 'glorot_uniform', activation = 'sigmoid'))
```

```python
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy',metrics=['accuracy'])
```

```python
classifier.summary()
```

```python
model_history=classifier.fit(X_train,y_train,batch_size=10,epochs=30,validation_split=0.25)
```

```python
import time
import os

def saveModel_path(model_dir="SAVED_MODELS"):
    os.makedirs(model_dir, exist_ok=True)
    fileName = time.strftime("Model_%Y_%m_%d_%H_%M_%S_.h5")
    model_path = os.path.join(model_dir, fileName)
    print(f"your model will be saved at the following location\n{model_path}")
    return model_path
```

# Weight initialization is a crucial step in training Artificial Neural Networks (ANNs) because it significantly impacts the convergence speed and the ability to reach a good solution. Here are some of the most common weight initialization techniques:

# 1. Zero Initialization
## Description: All weights are initialized to zero.
Pros: Simple and easy to implement.
Cons: This leads to symmetry, where neurons in the network learn the same features, effectively making the network incapable of learning. This is why zero initialization is not used in practice.

# 2. Random Initialization
## Description: Weights are initialized randomly from a small Gaussian distribution (mean = 0, small standard deviation).
## Pros: Breaks symmetry, allowing the network to learn different features.
## Cons: If the variance is not controlled, it can lead to vanishing or exploding gradients, especially in deep networks.

3. Xavier/Glorot Initialization
### Xavier/Glorot Initialization Formula

For a layer with $$ \( n_{\text{in}} \) $$ input units and $$ \( n_{\text{out}} \) $$ output units, the Xavier/Glorot initialization is defined as:

1. **Normal Distribution**: $$
   \[
   W \sim \mathcal{N}\left(0, \frac{2}{n_{\text{in}} + n_{\text{out}}}\right)
   \] $$

2. **Uniform Distribution**: $$
   \[
   W \sim U\left[-\sqrt{\frac{6}{n_{\text{in}} + n_{\text{out}}}}, \sqrt{\frac{6}{n_{\text{in}} + n_{\text{out}}}}\right]
   \] $$




# Pros: Works well for networks with sigmoid or hyperbolic tangent (tanh) activation functions, helping to keep the gradients stable.
# Cons: May not perform well with ReLU activation functions, as it does not account for the asymmetry of ReLU.

### 4. He Initialization Formula

Description: Similar to Xavier initialization but designed for ReLU activation functions. Weights are initialized from a Gaussian distribution with a variance of 2/n(input)

For a layer with $$ \( n_{\text{in}} \) $$ input units (neurons):

1. **Normal Distribution**: $$
   \[
   W \sim \mathcal{N}\left(0, \frac{2}{n_{\text{in}}}\right)
   \] $$

2. **Uniform Distribution**: $$
    \[
   W \sim U\left[-\sqrt{\frac{6}{n_{\text{in}}}}, \sqrt{\frac{6}{n_{\text{in}}}}\right]
   \]
   $$

# Pros: Specifically tailored for ReLU and its variants, helping to prevent vanishing/exploding gradients.
# Cons: Can sometimes still suffer from exploding gradients in very deep networks.
   


# 5. Lecun Initialization
Description: Weights are initialized from a Gaussian distribution with a variance of 1/n(input)

## This is specifically designed for activation functions like the sigmoid.
Pros: Well-suited for networks with sigmoid or softmax activation functions.
Cons: Not ideal for ReLU, which tends to require larger initial weights

# 6. Orthogonal Initialization
Description: Weights are initialized with an orthogonal matrix. This ensures that the weights are not only randomly initialized but also independent.
Pros: Helps maintain the variance of activations across layers.
Cons: More complex to implement and compute, generally used in recurrent neural networks (RNNs) rather than feedforward networks.

# 7. Variance Scaling Initialization
Description: A generalization of several initialization schemes. Weights are initialized from a Gaussian distribution scaled by a factor determined by the chosen mode (fan_in, fan_out, or fan_avg).
Pros: Can be adapted to different network architectures and activation functions.
Cons: Requires careful selection of the mode for best performance.

# Summary
* Xavier/Glorot Initialization is suitable for sigmoid and tanh activations.
* He Initialization is best for ReLU and its variants.
* Lecun Initialization works well with sigmoid and softmax.
* Orthogonal Initialization is more specialized and often used in RNNs.
* Variance Scaling provides flexibility depending on the architecture.

##Choosing the right weight initialization technique can significantly improve the performance and stability of your neural network, especially as it becomes deeper and more complex.

# Building a Model with Different Weight Initialization Techniques

```python
from keras.models import Sequential
from keras.layers import Dense
from keras import initializers

# Initialize a Sequential model
model = Sequential()

# Input Layer with Xavier/Glorot Initialization
model.add(Dense(units=64,
                input_shape=(100,),
                kernel_initializer=initializers.glorot_uniform(),
                activation='relu'))

# Hidden Layer with He Initialization
model.add(Dense(units=64,
                kernel_initializer=initializers.he_normal(),
                activation='relu'))

# Hidden Layer with Lecun Initialization
model.add(Dense(units=64,
                kernel_initializer=initializers.lecun_normal(),
                activation='relu'))

# Hidden Layer with Orthogonal Initialization
model.add(Dense(units=64,
                kernel_initializer=initializers.orthogonal(),
                activation='relu'))

# Output Layer with Variance Scaling Initialization
model.add(Dense(units=10,
                kernel_initializer=initializers.VarianceScaling(scale=2.0, mode='fan_in', distribution='normal'),
                activation='softmax'))

# Compile the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Model Summary
model.summary()
```
